```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', marker='o')

# Add title and labels
plt.title('Sine Wave Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
